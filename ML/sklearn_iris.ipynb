iris.feature_names 칼럼으로 iris 데이터프레임 생성

```python
import pandas as pd

iris_pd = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_pd.head()
```

iris.target은 iris 품종, 새로운 칼럼인 species에 추가

```python
iris_pd['species'] = iris.target
iris_pd.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

sepal length boxplot

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="sepal length (cm)", y="species", data=iris_pd, orient='h', palette = "Set2")
```

sepal width box plot

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="sepal width (cm)", y="species", data=iris_pd, orient='h', palette='Set2')
```

petal length box plot

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="petal length (cm)", y="species", data=iris_pd, orient='h', palette="Set2")
```

petal width boxplot

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="petal width (cm)", y="species", data=iris_pd, orient='h', palette="Set2")
```

iris[species] != 0 인 iris 데이터들을 따로 모음 (iris_12)

```python
iris_12 = iris_pd[iris_pd['species']!=0]
iris_12.info()
```

위에서 sepal로는 구분하기 어려운 걸 확인했으니 petal에 집중

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x="petal length (cm)", y="petal width (cm)", data=iris_12, hue="species", palette="Set2")
```

Decision Tree

0: sepal length, 1: sepal width, 2: petal length, 3: petal width 이므로   
iris.data[:, 2:] 는 모든 데이터들에 대하여 petal length, width를 추출하는 코드임

```python
iris.data[:, 2:]
```

```python
from sklearn.tree import DecisionTreeClassifier

iris_tree = DecisionTreeClassifier()
iris_tree.fit(iris.data[:, 2:], iris.target)
```

```python
from sklearn.metrics import accuracy_score

y_pred_tr = iris_tree.predict(iris.data[:, 2:])
accuracy_score(iris.target, y_pred_tr)
```
